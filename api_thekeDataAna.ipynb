```python
### fetching all page data from the tmdb website using api

import requests

TOKEN = "bearer token/api read access token from tmdb"  ###api read access token from tmdb
headers = {
    "Authorization": f"Bearer {TOKEN}",
    "accept": "application/json"
}

url = "https://api.themoviedb.org/3/keyword/1721/movies"

response = requests.get(url, headers=headers)
data = response.json()

print("Page no: ",data["page"])
print("Total Pages: ",data["total_pages"])
print("Total Results: ",data["total_results"])
```

```python
###saving info of movies in movies variable

movies = []

for page in range(1, data["total_pages"] + 1):
    response = requests.get(
        url,
        headers=headers,
        params={"page": page}
    )

    page_data = response.json()
    movies.extend(page_data["results"])

print(f"Collected {len(movies)} movies")
```

```python
### all data extfrom

import pandas as pd

df = pd.DataFrame(movies)

df.head()
```

```python
table = df[
    [
        "id",
        "title",
        "release_date",
        "vote_average",
        "vote_count",
        "popularity"
    ]
]

table.head()
```

```python
table.to_csv("tmdb_movies.csv", index=False)
```
